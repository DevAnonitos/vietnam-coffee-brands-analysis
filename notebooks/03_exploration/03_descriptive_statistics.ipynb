# IV. Descriptive Statistics

## IV.A. Measures of Central Tendency

```python
# summary statistical

df.describe(include="all")
```

### IV.A.1. Mean

```python
# Mean

df.mean()
```

### IV.A.2. Median

```python
# Median

df.median()
```

### IV.A.3. Mode

```python
# Mode

df.mode()
```

## IV.B. Measures of Dispersion

### IV.B.1. Range

```python
# Range

df.max() - df.min()
```

### IV.B.2. Variance

```python
# Variance

df.var()
```

### IV.B.3. Standard Deviation

```python
# Standard Deviation

df.std()
```

### IV.B.4. Interquartile Range

```python
# Interquartile Range (IQR)

df.quantile(0.75) - df.quantile(0.25)
```

### IV.B.5. Coefficient of Variation

```python
# Coefficient of Variation (CV)

df.std() / df.mean()
```

## IV.C. Distribution Shape

### IV.C.1. Skewness

```python
# Skewness

df.skew()
```

### IV.C.2. Kurtosis

```python
# Kurtosis

df.kurt()
```

### IV.C.3. Percentiles

```python
# Percentiles

df.quantile([0.1, 0.25, 0.5, 0.75, 0.9])
```
