```python
def fetch_html(url: str) -> str:
    response = requests.get(url)
    response.raise_for_status()
    return response.text

html = fetch_html(url)
print(html)
```

```python
def parse_html(html: str) -> BeautifulSoup:
    soup = BeautifulSoup(html, "html.parser")
    return soup

soup = parse_html(html)
print(soup.prettify())
```
